```python
def DisplayBoard(board):
    print(("+-------")*3+"+")
    for i in range(3):
        print("|       "*3+"|")
        for j in range(3):
            print("|   "+ str(board[i][j])+"  ", end=" ")
        print("|")
        print("|       "*3+"|")
        print(("+-------")*3+"+")
```

```python
def EnterMove(board):
    ok= False
    while not ok:
        move=int(input("Enter a position number: "))
        move=move-1
        row=move//3
        col=move%3
        sign=board[row][col]
        ok=sign not in ["X", "O"]
        if not ok:
            print("Bad move!")
        if move>=0 and move<9 and ok:
            board[row][col]="O"
```

```python
def MakeListOfFreeFields(board):
    free=[]
    for row in range(3):
        for col in range(3):
            if board[row][col] not in ["X", "O"]:
                free.append(board[row][col])
    return free
```

```python
def VictoryFor(board,sign):
    if sign=="X":
        who='me'
    elif sign=="O":
        who='you'
    else:
        who=None
    for pos in range(3):
        if str(board[pos][0])==sign and str(board[pos][1])==sign and str(board[pos][2])==sign:
            return who
        if str(board[0][pos])==sign and str(board[1][pos])==sign and str(board[2][pos])==sign:
            return who
    if str(board[0][0])==sign and str(board[1][1])==sign and str(board[2][2]==sign):
        return who
    if str(board[2][0])==sign and str(board[1][1])==sign and str(board[0][2])==sign:
        return who
    return None
```

```python
import random
def DrawMove(board):
    free = MakeListOfFreeFields(board) # make a list of free fields
    cnt = len(free)
    if cnt > 0:	# if the list is not empty, choose a place for 'X' and set it
        mymove=random.choice(free)-1
    row=mymove//3
    col=mymove%3
    board[row][col]="X"
```

```python
board=[[3*j+i+1 for i in range(3)] for j in range(3)]
board[1][1]='X'
free=MakeListOfFreeFields(board)
DisplayBoard(board)
humanturn=True
while len(free)>0:
    if humanturn==True:
        EnterMove(board)
        victor=VictoryFor(board,"O")
    else:
        print("My Turn")
        DrawMove(board)
        victor=VictoryFor(board,"X")
    if victor!=None:
        break
    humanturn=not humanturn
    free = MakeListOfFreeFields(board)
    DisplayBoard(board)
DisplayBoard(board)
if victor=="you":
    print("You Won!")
elif victor=="me":
    print("I Won!")
else:
    print("Tie!")
```
